```python
# Read Planning Area population
popsheet = pd.read_excel("t1-9.xls", sheet_name="T7(Total)")

# Dropping empty rows
popdf = popsheet.dropna(how="all", thresh=2)
popdf = popdf.dropna(how="all", thresh=2, axis=1)

# Making header
popdf.columns = popdf.iloc[0]
popdf = popdf[1:]

# Removing redundant headers
popdf = popdf[popdf["Planning Area"] != "Planning Area"]

# Dropping Subzone data
popdf = popdf[popdf["Planning Area"] != "NaN"]

popdf.head()
```
